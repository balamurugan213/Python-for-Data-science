## Example 1

```python
from bs4 import BeautifulSoup
import requests

def get(URL):
    """
    returns the website data html content
    """
    return requests.get(url=URL).content

def extract_links(URL, content):
    """
    extracts all hyperlinks from the html recieved and returns a list of them
    :param content: html of the webpage
    :return: list of URLS
    """
    soup = BeautifulSoup(content, "lxml")
    links = []
    for link in soup.findAll('a'):
        # print(link)
        if link.get('href').startswith("/"):
            # links.append(link.get('href'))
            continue
        links.append(link.get('href'))
    return links

if __name__ == '__main__':
  URL = "http://flask.pocoo.org"
  print(extract_links(URL, get(URL)))
```

## Example 2

```python
import re
import requests

final_links = []
processed = []


def searchSite(url, depth=2, search="charger"):
    # only do http links
    if url.startswith("http://") or url.startswith("https://") and url not in processed:
        processed.append(url)
        # split out the url into host and doc
        host = url

        # make the first request
        # print("crawling URL: " + host)
        try:
            res = requests.get(host)
            contents = str(res.content)
            m = re.findall('href="(.*?)"', contents)

            if search in contents:
                # print("Found " + search + " at " + url)
                final_links.append(url)
            # print(str(depth) + ": processing " + str(len(m)) + " links")
            for href in m:
                # relative urls
                if href.startswith("/"):
                    href = host + href
                if depth > 0:
                    searchSite(href, depth - 1, search)
        except requests.exceptions.ConnectionError:
            print("cannot connect to", host)

    else:
        print("skipping " + url)


if __name__ == '__main__':
  searchSite(url="https://www.oneplus.in", depth=1, search="charger")
  print(final_links)
```

## *Example* 3
* Take this page -  https://en.wikipedia.org/wiki/Database
* Extract all the text using beautiful soup
* Create a DataFrame that contains word and its frequency
* Output the top 10 occurring words
* plot the word frequencies.

```python
import requests
from bs4 import BeautifulSoup
page = requests.get("https://writingexplained.org/").content
soup = BeautifulSoup(page,"lxml")
soup.text
```

```python
text = soup.text
freq = {}
for word in text.split():
  if word in freq:
    freq[word]= freq[word] + 1
  else:
    freq[word] = 1

freq
```
